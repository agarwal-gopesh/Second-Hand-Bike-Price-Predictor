## Column Description 

- `title` : Name and model of the bike listing, usually including brand and variant

- `price` : Listed selling price of the second-hand bike (typically in INR)

- `kilometer` : Total distance the bike has been ridden, usually in kilometers

- `fuel_type` : Type of fuel the bike uses (e.g., Petrol, Electric)

- `registration_type` : Type of vehicle registration (e.g., Individual, Commercial)

- `registration_year` : Year in which the bike was officially registered

- `owner` : Number/type of previous owners (e.g., First owner, Second owner)

- `color` : Color of the bike as specified in the listing

- `city` : Location (city) where the bike is available for sale

- `insurance` : Insurance status or validity of the bike

- `last_updated` : Timestamp indicating when the listing was last updated

- `mileage_arai` : Official mileage certified under standard testing conditions (ARAI)

- `mileage_owner_reported` : Mileage reported by actual users based on real-world usage

- `riding_range` : Maximum distance the bike can travel on a full tank/charge

- `cylinders` : Number of engine cylinders in the bike

- `front_brake_type` : Type of braking system used in the front wheel (e.g., Disc, Drum)

- `rear_brake_type` : Type of braking system used in the rear wheel

- `tyre_type` : Type of tyres used (e.g., Tubeless, Tube)

- `fuel_tank_capacity` : Total capacity of the fuel tank in liters

- `reserve_fuel_capacity` : Amount of fuel reserved for emergency use after main fuel is low

```python
bike_data['reg_year'] = bike_data['title'].str.extract(r'(\d{4})')
bike_data['reg_year'] = bike_data['reg_year'].astype(int)

brands = ["Harley-Davidson","Royal Enfield","Hero Honda","TVS Motor","Moto Guzzi","CFMoto","Husqvarna","Kawasaki","Triumph","Mahindra","Benelli","Keeway","Aprilia","Suzuki",
          "Yamaha","Bajaj","Honda","Hero","TVS","KTM","Jawa","Yezdi","BMW","Ducati","Indian","Ampere Magnus","OLA",
          "Ather","Okinawa","Revolt","Ampere","Ola","Vida","Joy e-bike","PURE EV","Evolet","Bounce","Odysse","Ultraviolette",
          "Vespa","UM","Benling","BSA"]

def extract_brand_name(title) : 
    for brand in brands :
        if brand.lower() in title.lower() :
            return brand 
    return "Unknown"

bike_data['brand_name'] = bike_data['title'].apply(extract_brand_name)  
bike_data1 = bike_data.copy()

bike_data1['temp'] = bike_data1.apply(lambda x: str(x['title']).replace(str(x['reg_year']), '').replace(str(x['brand_name']), ''), axis=1)
bike_data1['model'] = bike_data1['temp'].str.strip().str.replace(r'\s+', ' ', regex=True)
bike_data1.drop(columns=['temp'], inplace=True)

bike_data1.info()
```

```python
def convert_price(price) : 
    price = str(price)
    price = price.replace('Rs.', '').strip()

    if 'Lakh' in price :
        num = float(price.replace('Lakh', '').strip())
        return num * 100000
    else :
        return float(price.replace(',', '').strip())
    
bike_data1['price'] = bike_data1['price'].astype(str).apply(convert_price)
bike_data1.drop(columns=['registration_year'], inplace=True)
bike_data1.head(5)
```

```python
import numpy as np
bike_data1['fuel_tank_capacity'] = bike_data1['fuel_tank_capacity'].replace('Not Available' , np.nan)
bike_data1['fuel_tank_capacity'] = bike_data1['fuel_tank_capacity'].str.replace('litres', '').astype(float)

bike_data1['reserve_fuel_capacity'] = bike_data1['reserve_fuel_capacity'].replace('Not Available' , np.nan)
bike_data1['reserve_fuel_capacity'] = bike_data1['reserve_fuel_capacity'].str.replace('litres', '').astype(float)

bike_data1.head(6)
```

```python
# dropped duplicate rows 
bike_data1.drop([2742 , 1285 , 2352 , 2510 , 1143 , 2171 , 1865 , 2574]) 
```

```python
bike_data1['cylinders'] = bike_data1['cylinders'].replace('Not Available' , np.nan)
bike_data1['cylinders'] = bike_data1['cylinders'].astype(float)
bike_data1.info()
```
